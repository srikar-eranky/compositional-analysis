```python
cluster_size = []
composition_list = []
with open('clustering_4v_4v_baseline.species') as tf:
    lines = tf.readlines()
    #print(lines)

for line in lines:
    if '#' in line:
        pass
    else:
        cs = line.count('.') + 1
        #print('cluster size: ', cs)
        A4_count = line.count('A4')
        B4_count = line.count('B4')
        #print('A4: ', A4_count, 'B4: ', B4_count)
        if cs > 1:
            composition = A4_count/B4_count
            composition_list.append(composition)
            cluster_size.append(cs)
    #print(line)
    
```

```python
composition_list
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(cluster_size,composition_list)
plt.xlabel('Cluster Size')
plt.ylabel('Composition')
plt.title('1000:1000 ratio')
```
