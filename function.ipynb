```python
import matplotlib.pyplot as plt
import numpy as np
def ratio_calculator(species_file,plot_title,saveIt=False):
    cluster_size = []
    composition_list = []
    with open(species_file,'r') as tf:
        lines = tf.readlines()
        #print(lines)

    for line in lines:
        if '#' in line:
            pass
        else:
            cs = line.count('.') + 1
            #print('cluster size: ', cs)
            A4_count = line.count('A4')
            B4_count = line.count('B4')
            #print('A4: ', A4_count, 'B4: ', B4_count)
            if cs > 100:
                composition = A4_count/B4_count
                composition_list.append(composition)
                cluster_size.append(cs)
        #print(line)
    plt.figure(figsize=(10,6))    
    plt.scatter(cluster_size, composition_list)
    plt.xlabel('Cluster Size (Molecules)',fontsize=15)#add label_size
    plt.ylabel('Cluster Size Composition (A4:B4)',fontsize=15)
    plt.title(plot_title,fontsize=20)
    mean_ratio = np.mean(composition_list)
    plt.axhline(mean_ratio, ls='dashed', color='black', label=f'mean_ratio={mean_ratio:.3f}')
    plt.legend()
    if saveIt:
        plt.savefig(plot_title + '.png',dpi=200)
```
